```python

!pip install python-docx gradio

import gradio as gr
from google.colab import files
from docx import Document

print("Envie o arquivo para Advertência I:")
uploaded_i = files.upload()
arquivo_i = list(uploaded_i.keys())[0]

print("Envie o arquivo para Advertência II:")
uploaded_ii = files.upload()
arquivo_ii = list(uploaded_ii.keys())[0]

print("Envie o arquivo para Advertência III:")
uploaded_iii = files.upload()
arquivo_iii = list(uploaded_iii.keys())[0]


def preencher_documento(nome_arquivo, nome, num_contrato, email, dia, mes, numero, valor, vencimento,
                        numero_extra, valor_extra, vencimento_extra,
                        numero_extra2, valor_extra2, vencimento_extra2,
                        numero_extra3, valor_extra3, vencimento_extra3,
                        numero_extra4, valor_extra4, vencimento_extra4,
                        numero_extra5, valor_extra5, vencimento_extra5):
    documento = Document(nome_arquivo)

    dados_cliente = {
        "{{NOME}}": nome,
        "{{NºCONTRATO}}": num_contrato,
        "{{EMAIL}}": email,
        "{{DIA}}": dia,
        "{{MES}}": mes,
        "{{NÚMERO}}": numero,
        "{{VALOR}}": valor,
        "{{VENCIMENTO}}": vencimento
    }


    for paragrafo in documento.paragraphs:
        for chave, valor in dados_cliente.items():
            if chave in paragrafo.text:
                paragrafo.text = paragrafo.text.replace(chave, valor)

    for tabela in documento.tables:
        for linha in tabela.rows:
            for celula in linha.cells:
                for chave, valor in dados_cliente.items():
                    if chave in celula.text:
                        celula.text = celula.text.replace(chave, valor)


        campos_extra = [
            (numero_extra, valor_extra, vencimento_extra),
            (numero_extra2, valor_extra2, vencimento_extra2),
            (numero_extra3, valor_extra3, vencimento_extra3),
            (numero_extra4, valor_extra4, vencimento_extra4),
            (numero_extra5, valor_extra5, vencimento_extra5)
        ]

        for campo in campos_extra:
            if campo[0] or campo[1] or campo[2]:
                nova_linha = tabela.add_row()
                celulas = nova_linha.cells
                for i, dado in enumerate(campo):
                    if dado:
                        celulas[i].text = dado

    novo_arquivo = f"{nome_arquivo.split('.')[0]}_preenchido.docx"
    documento.save(novo_arquivo)

    return novo_arquivo


def criar_interface(tipo, arquivo):
    with gr.Blocks() as interface:
        gr.Markdown(f"## 📄 Formulário - {tipo}")
        with gr.Row():
            with gr.Column():
                nome = gr.Textbox(label="👤 Nome")
                num_contrato = gr.Textbox(label="📜 Nº Contrato")
                email = gr.Textbox(label="📧 E-mail")
                dia = gr.Textbox(label="📅 Dia")
                mes = gr.Textbox(label="🗓️ Mês")

            with gr.Column():
                numero = gr.Textbox(label="🔢 Número Da fatura")
                valor = gr.Textbox(label="💰 Valor")
                vencimento = gr.Textbox(label="⏳ Vencimento")

                show_extra = gr.Checkbox(label="Adicionar campos extras (Número2, Valor2, Vencimento2 até Número5, Valor5, Vencimento5)", value=False)


                with gr.Row(visible=False) as extra_fields:
                    numero_extra = gr.Textbox(label="➕ Número2 (Opcional)", placeholder="Opcional")
                    valor_extra = gr.Textbox(label="💵 Valor2 (Opcional)", placeholder="Opcional")
                    vencimento_extra = gr.Textbox(label="🕒 Vencimento2 (Opcional)", placeholder="Opcional")
                    numero_extra2 = gr.Textbox(label="➕ Número3 (Opcional)", placeholder="Opcional")
                    valor_extra2 = gr.Textbox(label="💵 Valor3 (Opcional)", placeholder="Opcional")
                    vencimento_extra2 = gr.Textbox(label="🕒 Vencimento3 (Opcional)", placeholder="Opcional")
                    numero_extra3 = gr.Textbox(label="➕ Número4 (Opcional)", placeholder="Opcional")
                    valor_extra3 = gr.Textbox(label="💵 Valor4 (Opcional)", placeholder="Opcional")
                    vencimento_extra3 = gr.Textbox(label="🕒 Vencimento4 (Opcional)", placeholder="Opcional")
                    numero_extra4 = gr.Textbox(label="➕ Número5 (Opcional)", placeholder="Opcional")
                    valor_extra4 = gr.Textbox(label="💵 Valor5 (Opcional)", placeholder="Opcional")
                    vencimento_extra4 = gr.Textbox(label="🕒 Vencimento5 (Opcional)", placeholder="Opcional")
                    numero_extra5 = gr.Textbox(label="➕ Número6 (Opcional)", placeholder="Opcional")
                    valor_extra5 = gr.Textbox(label="💵 Valor6 (Opcional)", placeholder="Opcional")
                    vencimento_extra5 = gr.Textbox(label="🕒 Vencimento6 (Opcional)", placeholder="Opcional")

                show_extra.change(fn=lambda show: extra_fields.update(visible=show), inputs=[show_extra], outputs=[extra_fields])

        botao = gr.Button("📂 Gerar Documento", variant="primary")
        saida = gr.File(label="📥 Documento Preenchido")

        botao.click(fn=lambda *args: preencher_documento(arquivo, *args),
                    inputs=[nome, num_contrato, email, dia, mes, numero, valor, vencimento,
                            numero_extra, valor_extra, vencimento_extra,
                            numero_extra2, valor_extra2, vencimento_extra2,
                            numero_extra3, valor_extra3, vencimento_extra3,
                            numero_extra4, valor_extra4, vencimento_extra4,
                            numero_extra5, valor_extra5, vencimento_extra5],
                    outputs=saida)

    return interface


tabs = gr.TabbedInterface(
    [
        criar_interface("Advertência I", arquivo_i),
        criar_interface("Advertência II", arquivo_ii),
        criar_interface("Termo de eliminação COOP", arquivo_iii)
    ],
    ["⚠️ Advertência I", "⚠️ Advertência II", "⚠️ Advertência III"]
)


tabs.launch()
```
